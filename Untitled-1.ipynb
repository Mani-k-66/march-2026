```python
class Student:
    def __init__(self,id,name,marks,greade):
        self.id="id"
        self.name="name"
        self.marks="marks"
        self.greade="greade"
```

```python
def display(self):
    print(f"id:{self.id},name:{self.name},marks:{self.marks},greade:{self.greade}")
```

```python
class StudentSystem:
    def __init__(self):
        self.students=[]
```

```python
def add_student(self):
    id=input("enter id: ")
    name=input('enter name:')
    marks=input("enter marks:")
    gread=float(input("enter greade:"))
    student=Student(id,name,marks,greade)
    self.students.append(student)
    print("student added successfully")
    
```

```python
def show_students(self):
    if not self.students:
        print("student not found")
    else:
        for student in self.students:
            student.display()
```

```python
def search_student(self):
    id=input("enter id to search:")
    for student in self.students:
        if student.id==id:
            student.display()
            return
        print("student not found")
```

```python
system=StudentSystem()
system.add_student()
system.show_students()
system.search_student()
```

```python
while True:
    print("\n1.Add  2.Show  3.Search  4.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        system.add_student()
    elif choice == "2":
        system.show_students()
    elif choice == "3":
        system.search_student()
    elif choice == "4":
        break
    else:
         print("invalid choice.please try again")
```
